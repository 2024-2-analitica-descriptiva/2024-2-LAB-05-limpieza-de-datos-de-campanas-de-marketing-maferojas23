```python

# Ruta de la carpeta que contiene los archivos .csv.zip
carpeta = "../files/input"

# Lista para almacenar los DataFrames leídos
dataframes = []

# Iterar sobre los archivos en la carpeta
for archivo in os.listdir(carpeta):
    if archivo.endswith(".csv.zip"):
        ruta_zip = os.path.join(carpeta, archivo)
        
        # Abrir el archivo .zip
        with ZipFile(ruta_zip) as zip_file:
            # Encontrar el archivo .csv dentro del .zip
            for nombre_archivo in zip_file.namelist():
                if nombre_archivo.endswith(".csv"):
                    # Leer el archivo .csv directamente desde el .zip
                    with zip_file.open(nombre_archivo) as csv_file:
                        df = pd.read_csv(csv_file)
                        dataframes.append(df)

# Combinar todos los DataFrames en uno solo
data_completa = pd.concat(dataframes, ignore_index=True)
print(data_completa)
```
